```python
## Env alias - mlcw

## Utilities
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
crop_losses_df = pd.read_csv('my_data/crop_losses_data.csv')
export_data_df = pd.read_csv('my_data/export_data_target.csv')
export_quantity_df = pd.read_csv('my_data/export_quantity_data.csv')
food_inflation_df = pd.read_csv('my_data/food_inflation_data.csv')
import_data_value_df = pd.read_csv('my_data/import_data_value.csv')
import_quantity_df = pd.read_csv('my_data/import_quantity_data.csv')
exchange_data_df = pd.read_csv('my_data/exchange_data.csv')
agricultural_land_df = pd.read_csv('my_data/agricultural_land.csv')
agricultural_permanent_crops_df = pd.read_csv('my_data/agricultural_permanent_crops.csv')
foreign_investment_total_inflows_df = pd.read_csv('my_data/fdi_total_inflows.csv')
foreign_investment_total_outflows_df = pd.read_csv('my_data/fdi_total_outflows.csv')
food_security_political_stability_df = pd.read_csv('my_data/political_stability.csv')

# Select and rename the 'Value' column in each Dataframe
agricultural_land_df = agricultural_land_df[['Area', 'Year', 'Value']].rename(columns={'Value': 'Agricultural Land'})
agricultural_permanent_crops_df = agricultural_permanent_crops_df[['Area', 'Year', 'Value']].rename(columns={'Value': 'Permanent Crops Area'})
crop_losses_df = crop_losses_df[['Area', 'Year', 'Value']].rename(columns={'Value': 'Crop Losses'})
exchange_data_df = exchange_data_df[['Area', 'Year', 'Value']].rename(columns={'Value': 'Exchange Data'})
export_data_df = export_data_df[['Area', 'Year', 'Value']].rename(columns={'Value': 'Export Value'})
export_quantity_df = export_quantity_df[['Area', 'Year', 'Value']].rename(columns={'Value': 'Export Quantity'})
foreign_investment_total_inflows_df = foreign_investment_total_inflows_df[['Area', 'Year', 'Value']].rename(columns={'Value': 'Total FDI inflow'})
foreign_investment_total_outflows_df = foreign_investment_total_outflows_df[['Area', 'Year', 'Value']].rename(columns={'Value': 'Total FDI outflow'})
food_inflation_df = food_inflation_df[['Area', 'Year', 'Value']].rename(columns={'Value': 'Food Inflation'})
import_data_value_df = import_data_value_df[['Area', 'Year', 'Value']].rename(columns={'Value': 'Import Value'})
import_quantity_df = import_quantity_df[['Area', 'Year', 'Value']].rename(columns={'Value': 'Import Quantity'})
food_security_political_stability_df = food_security_political_stability_df[['Area', 'Year', 'Value']].rename(columns={'Value': 'Political Stability Index'})

# Merge all datasets on 'Area' and 'Year' using an outer join
from functools import reduce
data_frames = [import_quantity_df, export_quantity_df, import_data_value_df, food_inflation_df, exchange_data_df, crop_losses_df, 
               agricultural_land_df, agricultural_permanent_crops_df, food_security_political_stability_df, 
               foreign_investment_total_inflows_df, foreign_investment_total_outflows_df, export_data_df]
merged_df = reduce(lambda left, right: pd.merge(left, right, on=['Area', 'Year'], how='outer'), data_frames)

# Handling missing values
# Fill missing values with a placeholder like zero or median
merged_df.fillna(0, inplace=True)

# Save the merged dataframe
merged_df.to_csv('my_data/final_dataset_test.csv', index=False)

merged_df
```
